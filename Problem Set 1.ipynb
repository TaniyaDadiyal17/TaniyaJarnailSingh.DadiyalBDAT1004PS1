```python
using System;

namespace ConsoleApp5
{
    class Program
    {
        static void Main(string[] args)
        {
            Console.WriteLine("Input the number of elements to be stored in the array :5");
            int read_num = Convert.ToInt32(Console.ReadLine());
            int[] array_input = new int[read_num];
            for (int i = 0 ; i < read_num; i++)
            {
                Console.WriteLine("element - " + i + " : ");
                array_input[i] = Convert.ToInt32(Console.ReadLine());
            }
            Console.WriteLine("The Even elements are:");
            for (int i = 0; i < read_num; i++)
            {
                if (array_input[i] % 2 == 0)
                {
                    Console.WriteLine(array_input[i]);
                }
            }
            Console.WriteLine("The Odd elements are:");
            for (int i = 0; i < read_num; i++)
            {
                if (array_input[i] % 2 != 0)
                {
                    Console.WriteLine(array_input[i]);
                }
            }
        }
    }
}
```

```python
def pig(word):
    first_letter = word[0]
    rest_of_word = word[1 : ]
    if first_letter in('a','e','i','o','u'):
        print (first_letter + rest_of_word + "way")
    else:
         print (rest_of_word + first_letter+ "ay")
```

```python
pig('Farm')
pig('October')
pig('beautiful')
pig('garden')
pig('chicken')
pig('mud')
pig('fruits')
pig('house')
```

# Question 10 

```python
def frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
print(frequency('apple'))
print(frequency('beautiful'))
print(frequency('farm'))
print(frequency('happy'))
print(frequency('country'))
print(frequency('house'))
```
